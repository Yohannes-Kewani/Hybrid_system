**Introduction**

Although Ethiopia is consistently expanding its national power generation capabilities, a significant disparity remains in electricity access, particularly among unserved communities. Statistical data reveals that approximately 80% to 85% of the Ethiopian population resides in rural hinterlands, where the infrastructure required for grid connectivity is currently nonexistent. Due to the dispersed settlement patterns of these rural homesteads, extending the national grid to reach them is not economically feasible in the short to medium term. To overcome this bottleneck in energy delivery, harnessing the country's abundant untapped renewable potentials—specifically wind and solar resources—is essential for resolving the energy crisis in rural communities and enhancing their socio-economic well-being.

For this hybrid micro-grid design and performance analysis, the selected site is the rural vicinity of Mekelle City, located approximately 780 km north of the capital, Addis Ababa. The Mekelle area is well-identified for its high solar radiation intensity and significant wind energy potential. This potential is evidenced by the presence of the Ashegoda Wind Farm in the immediate vicinity. However, much like other large-scale projects, the Ashegoda station feeds its generated power directly into the National Grid (ICS), leaving the surrounding rural villages without access to the energy generated in their own backyard.

Therefore, the primary objective of this study is to design and conduct a performance analysis of an isolated hybrid Micro-Grid power supply system using HOMER Pro. The aim is to deliver a reliable and cost-effective electric supply to a rural village near Mekelle. The proposed hybrid system integrates Wind Turbine Generators, Diesel Generators, and Solar PV panels. A distinct feature of this specific design analysis is the intentional exclusion of battery storage or other energy storage technologies. This strategy has been adopted to significantly minimize the system's Net Present Cost (NPC) and initial capital expenditure. **1.1 Current and Forecasted Load Profile**

The electrical load data for this study was established through a combination of direct on-site field surveys and empirical analysis of the target rural village. The assessment defines the current AC load baseline and projects demand growth over 10-year and 15-year horizons.

The design scope encompasses 100 households as the primary residential load. In addition to domestic consumption, the micro-grid is sized to support a significant agricultural load: two AC water pumps, each with a rated capacity of 100 kW, utilized for community irrigation. The operational schedule for these pumps is distinct, limited to a two-hour duration in the morning and a two-hour duration in the evening.

The baseline load data was input into the HOMER software using January as the representative peak month. The resulting generated profiles—Daily, Seasonal, and Yearly—are illustrated in Figures 2, 3, and 4 respectively. To simulate realistic grid fluctuations, a day-to-day random variability of 10% and a time-step variability of 20% were applied to the model.

According to the simulation outputs, the system demonstrates a scaled annual average energy consumption of 1,845 kWh/day , corresponding to an average continuous power demand of 76.87 kW . As depicted in the Daily Profile, the load is characterized by two distinct high-demand interval occurring from 00:00 to 03:00 and 12:00 to 15:00 , where the power usage spikes to a maximum. Consequently, the peak power requirement is approximately 250 kW , resulting in a system load factor of 0.31 . All power demands within the micro-grid are modeled as AC loads.

**2 LITERATURE Review**

Global electricity consumption continues to escalate, creating an urgent need for sustainable power solutions as fossil fuel reserves diminish. To mitigate reliance on conventional energy and address the inherent intermittency of standalone renewable technologies—such as solar photovoltaic (PV) or wind turbines—researchers increasingly advocate for hybrid off-grid systems. For instance, the design and performance analysis conducted for a rural village in the Adama area, Ethiopia [1], utilizes a hybrid approach to overcome resource variability, demonstrating the necessity of integrating multiple generation sources for reliable rural electrification.

To effectively engineer these complex systems, the Hybrid Optimization of Multiple Energy Resources (HOMER) software serves as a critical tool. HOMER simplifies the evaluation of both off-grid and grid-connected configurations by modeling various inputs, including conventional generators, wind turbines, solar PV, hydropower, batteries, and biomass [2]. As highlighted in the methodology of [1], the software performs a rigorous economic analysis, calculating the total Net Present Cost (NPC), operational costs, and the Levelized Cost of Energy (COE) to identify the optimal system architecture [3]. Ultimately, these simulations support the development of robust micro-grids—localized power networks that aggregate distributed generation resources (DGR) ranging from solar arrays and wind converters to biodiesel generators and micro-turbines [4].
